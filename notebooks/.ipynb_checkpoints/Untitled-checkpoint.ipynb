```python
from pyspark.sql.functions import col, count, when, isnan

# إجمالي الصفوف
total_rows = df.count()

# تحليل النسب المئوية للـ Null في residential_units لكل فئة
columns_to_check = ['building_class_category', 'building_class_at_present', 'borough_name', 'source_year']

for col_name in columns_to_check:
    print(f"\n==== Null ratios by {col_name} ====")
    df.groupBy(col_name) \
      .agg(
          count("*").alias("total_count"),
          count(when(col("residential_units").isNull(), True)).alias("null_count")
      ) \
      .withColumn("null_ratio", col("null_count") / col("total_count")) \
      .orderBy(col("null_ratio").desc()) \
      .show(truncate=False)
```

الأعمدة فيها null لأن بعض العقارات مش سكنية (زي مكاتب أو جراجات)
